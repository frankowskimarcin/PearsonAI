## Task1

```python
import pandas as pd

test_file = pd.read_csv('test.csv', sep=';')
test_level_file = pd.read_csv('test_level.csv', sep=';')
class_file = pd.read_csv('class.csv', sep=';')
class_file.head(10)
```

## Task2

```python
class_file2=class_file[class_file.teaching_hours != "15+"]
class_file2=class_file2.dropna(how="all")
class_file2.head(10)
```

## Task3

```python
df1 = class_file2[['id', 'name', 'teaching_hours']]
df1
```

```python
df2 = test_file[['id','class_id', 'created_at', 'authorized_at', 'test_level_id']]
df2 = df2.dropna(subset=['authorized_at'])
df2
```

```python
new_df = pd.merge(left=df1, right=df2, left_on='id', right_on='class_id')
new_df = new_df.drop(columns = ['class_id'])
new_df.rename(columns={'id_x':'class_id',
                       'name':'class_name',
                       'id_y':'test_id',
                       'created_at':'test_created_at',
                       'authorized_at':'test_authorized_at', 
                       'test_level_id':'test_level'}, inplace=True)
new_df
```

```python
new_df['class_test_number'] = new_df.groupby('class_id').cumcount() + 1
new_df
```

```python
new_df.to_csv('test_utilization.csv', index=False, sep=';')
```

## Task4

```python
df_test = test_file.loc[test_file['test_status'] == 'SCORING_SCORED'].dropna(subset=['authorized_at'])
df_test = df_test[['class_id', 'created_at', 'authorized_at', 'overall_score']]
df_test
```

```python
df_class = class_file2[['id','name','teaching_hours']]
df_class
```

```python
new_df2 = pd.merge(left=df_class, right=df_test, left_on='id', right_on='class_id')
new_df2 = new_df2.drop(columns = ['class_id'])
new_df2.rename(columns={'id':'class_id',
                       'name':'class_name',
                       'created_at':'test_created_at',
                       'authorized_at':'test_authorized_at'}, inplace=True)
new_df2 = new_df2.dropna(subset =['overall_score'])
new_df2
```

```python
new_df3 = new_df2.groupby(['class_id'])['overall_score'].mean().to_frame()
final_df = pd.merge(left=new_df2, right=new_df3, left_on='class_id', right_on='class_id')
final_df = final_df.drop(columns='overall_score_x')
final_df.rename(columns={
    'overall_score_y':'avg_class_test_overall_score'}, inplace=True)
final_df = final_df.drop_duplicates(subset=['class_id'])
final_df
```

```python
final_df.to_csv('test_average_scores.csv', index=False, sep=';')
```
